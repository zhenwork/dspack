```python
import matplotlib.pyplot as plt
```

```python
dnames = ["WT-1", "WT-2", "WT-3", "G150A-1", "G150A-2", "G150A-3", "G150T-1", "G150T-2", "G150T-3"]
modes = ["Protein_Refmac5","Solvent_Refmac5","Protein_PHENIX","Solvent_PHENIX"]
colors = ["tab:purple","tab:blue","tab:orange","tab:green"]
```

```python
plt.figure(figsize=(15,9))
for idx,dname in enumerate(dnames):
    plt.subplot(3,3,idx+1)
    for jdx,mode in enumerate(modes):
        plt.plot(data[dname]["Anisotropy"], data[dname][mode],"-o", color=colors[jdx],label=mode.replace("_"," "))
    plt.tick_params('y', length=10, which='major')
    plt.tick_params('x', length=5, which='major')
    plt.title(dname,fontsize=20)
    plt.legend(loc=1,fontsize=13)
    if idx>5:
        plt.xlabel("Anisotropy",fontsize=20)
    if idx%3==0:
        plt.ylabel("Fraction of atoms",fontsize=20)
    plt.xticks(np.arange(0,1.1,0.2), np.around(np.arange(0,1.1,0.2),1), fontsize=20)
    plt.yticks(np.arange(0,0.6,0.1),np.around(np.arange(0,0.6,0.1),1),fontsize=20)
    plt.xlim(0,1)
    plt.tight_layout()
    
fig = plt.gcf()
max_fit_dpi = min(int(np.sqrt(25.*1024**2/np.prod(fig.get_size_inches()))),600)
print "dpi = ", max_fit_dpi
plt.tight_layout()
plt.savefig("./600dpi/Fig_S01_Bragg_Models_Anisotropy_Distribution.jpg",dpi=600,bbox_inches='tight',pad_inches=0,transparent=True)
plt.savefig("./fitgdoc/Fig_S01_Bragg_Models_Anisotropy_Distribution.jpg",dpi=max_fit_dpi,bbox_inches='tight',pad_inches=0,transparent=True)
plt.show()
```
